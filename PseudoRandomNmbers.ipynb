```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng(100)

I = 1000

S = rng.normal(loc=100, scale=20, size=I)
S[:14]

S_mean = S.mean()
print(S_mean)


#A Historgram to Demonstrate My Data's Normal Distribution
plt.hist(S, density=True,  color='#a05195', bins=30)  # density=False would make counts
plt.axvline(x = S_mean, linestyle='--', color='#2f4b7c')
plt.ylabel('Probability')
plt.xlabel('Data')
```

```python
P = rng.random(I)
print(P[:10])

#A hisogram demonstrating the flat probability distribution:
plt.hist(P, density=True,  color='#a05195', bins=30)  # density=False would make counts
plt.axvline(x = P.mean(), linestyle='--', color='#2f4b7c')
plt.ylabel('Probability')
plt.xlabel('Data')

print(P.mean())
print(P.sum())

print(P[:10])

P /= P.sum() #Normalizes the values so that they sum up to one
print(P[:10])


np.dot(P, S)

```
