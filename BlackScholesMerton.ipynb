```python
import math
import pylab as py
import numpy as np

from pylab import mpl, plt

from numpy.random import default_rng
```

```python
#Some initial parameters:
S0 = 100
r = 0.05

sigma = 0.2
T = 1.0 #The time horizon in year fractions
I = 10000000

rng = default_rng(100)

ST = S0 * np.exp((r - sigma**2 / 2)*T + sigma*math.sqrt(T)*rng.standard_normal(I))

print(ST[:8].round(1))

print(ST.mean())

print(S0 * math.exp(r*T))
print(S0 * (1+r))

plt.style.use('seaborn')
mpl.rcParams['savefig.dpi'] = 300
mpl.rcParams['font.family'] = 'serif'

plt.figure(figsize=(15, 9))
plt.hist(ST, bins=35, label='frequency', color='#2f4b7c');

plt.axvline(ST.mean(), color='#bc5090', label='mean')
plt.axvline(ST.mean() + ST.std(), color='#ffa600', label='sd up')
plt.axvline(ST.mean() - ST.std(), color='#ffa600', label='sd down')

plt.legend(loc=0)
plt.xlabel('Monte-Carlo Estimate of the Call-Option Price.')
plt.ylabel('Frequency.');

```
